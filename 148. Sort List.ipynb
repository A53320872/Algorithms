## Method1: Don't create new nodes when merge: Space: O(1)

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def sortList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        def sortLN(head):
            if head == None or head.next == None: return head
            
            slow = head
            fast = head.next
            
            while fast and fast.next:
                slow = slow.next
                fast = fast.next.next
            
            mid = slow.next
            slow.next = None
            
            l = sortLN(head)
            r = sortLN(mid)
            
            return merge(l, r)
        def merge(l, r):
            
            result = ListNode(-1)
            ohead = result
            while l and r:
                if l.val < r.val:
                    result.next = l
                    result = result.next
                    l = l.next
                else:
                    result.next = r
                    result = result.next
                    r = r.next
            if l:
                result.next = l
            if r:
                result.next = r
            
            return ohead.next
        return sortLN(head)
```

## Method2: Create new nodes when merge: Space: O(n)

## Method3: transfer to array  Space: O(n)

```python
class Solution(object):
    def sortList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        nums = []
        original = head
        while head:
            nums.append(head.val)
            head = head.next
        if len(nums) <= 1: return original
        
        #print(nums)

        def sortArray(nums):
            if len(nums) == 1: return nums
            mid = len(nums) // 2
            l = nums[:mid]
            r = nums[mid:]
            
            l1 = sortArray(l)
            r1 = sortArray(r)
            
            return merge(nums, l1, r1)
        
        def merge(nums, l, r):
            i,j,k = 0,0,0
            while i < len(l) and j < len(r):
                if l[i] < r[j]:
                    nums[k] = l[i]
                    k += 1
                    i += 1
                else:
                    nums[k] = r[j]
                    k += 1
                    j += 1
            while i < len(l):
                nums[k] = l[i]
                k += 1
                i += 1
            while j < len(r):
                nums[k] = r[j]
                k += 1
                j += 1
            return nums
        
        nums = sortArray(nums)
        head = ListNode(nums[0])
        markHead = head
        for i in range(1, len(nums)):
            new = ListNode(nums[i])
            head.next = new
            head = head.next
        return markHead
        
        
```
