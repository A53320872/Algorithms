```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def deleteNode(self, root, key):
        """
        :type root: TreeNode
        :type key: int
        :rtype: TreeNode
        """
        
        delete = [key]
        pa = {root:None}
        head = [root]
        
        def findParent(root):
            if root == None: return
            if root.left != None:
                pa[root.left] = root
                findParent(root.left)
            if root.right != None:
                pa[root.right] = root
                findParent(root.right)
        findParent(root)
        
        def findDelete(root):
            if root == None: return 
            if root.val < delete[0]: findDelete(root.right)
            elif root.val > delete[0]: findDelete(root.left)
            else:
                # leaf
                if root.left == None and root.right == None:
                    if pa[root] != None:
                        if root == pa[root].left: pa[root].left = None
                        else: pa[root].right = None
                    else: head[0] = None
                    
                
                # only has left
                elif root.left != None and root.right == None:
                    if pa[root] != None:
                        if root == pa[root].left: pa[root].left = root.left
                        else: pa[root].right = root.left
                    else: head[0] = root.left
                
                # only has right
                elif root.right != None and root.left == None:
                    if pa[root] != None:
                        if root == pa[root].left: pa[root].left = root.right
                        else: pa[root].right = root.right
                    else: head[0] = root.right
                
                else:
                    # has left & right
                    minNode = root.right
                    while minNode != None:
                        if minNode.left == None: break
                        minNode = minNode.left
                    
                    if minNode == root.right:
                        root.val = minNode.val
                        root.right = minNode.right
                    else:
                        root.val = minNode.val
                        pa[minNode].left = minNode.right
                        
            
        findDelete(root)
        
        return head[0]
```
