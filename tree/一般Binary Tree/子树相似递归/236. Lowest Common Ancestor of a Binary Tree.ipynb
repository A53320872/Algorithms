```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def lowestCommonAncestor(self, root, p, q):
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """
        # method 1: using parent pointer
        pa = {root: None}
        def postOrder(root):
            if root == None:
                return
            postOrder(root.left)
            postOrder(root.right)
            if root.left != None:
                pa[root.left] = root
            if root.right != None:
                pa[root.right] = root
            return
        postOrder(root)
        
        pathP = set([])
        while p:
            pathP.add(p)
            p = pa[p]
        while q:
            if q in pathP:
                return q
            q = pa[q]
```

```python
# method 2: using subtree similarity
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def lowestCommonAncestor(self, root, p, q):
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """
        res = [None]
        def postOrder(root, p, q):
            if root == None:
                return False
            
            l = postOrder(root.left, p, q)
            r = postOrder(root.right, p, q)
            
            state = False
            if root == p:
                if l == True or r == True:
                    res[0] = p
                    state = False
                else:
                    state =  True
            elif root == q:
                if l == True or r == True:
                    res[0] = q
                    state = False
                else:
                    state = True
            else:
                if l == True and r == True:
                    res[0] = root
                    state =  False
                elif l == True or r == True:
                    state = True
                else:
                    state = False
            return state
        postOrder(root, p, q)
        return res[0]
```
