```python
最大堆
维护一个大小为k的最大堆
时间复杂度nlogk
```

```python
class Solution(object):
    def maxSlidingWindow(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: List[int]
        """
        def swap(tree, i, max_):
            tmp = tree[i]
            tree[i] = tree[max_]
            tree[max_] = tmp

        def heapify(tree, n, i):
            if i >= n: return
            left_child = 2 * i + 1
            right_child = 2 * i + 2
            max_ = i
            if left_child < n and tree[left_child] > tree[max_]:
                max_ = left_child
            if right_child < n and tree[right_child] > tree[max_]:
                max_ = right_child
            if max_ != i:
                swap(tree, i, max_)
                heapify(tree, n, max_)

        def buildHeap(tree, n):
            last_node = n - 1
            parent = (last_node - 1) // 2
            i = parent
            while i >= 0:
                heapify(tree, n, i)
                i -= 1
            return tree


        def upMove(heap, n, i):
            if i <= 0: return
            parent = (i-1) // 2
            if heap[i] > heap[parent]:
                swap(heap, i, parent)
            upMove(heap, n, parent)

        def heapInsert(heap, n, item):
            heap.append(item)
            upMove(heap, n+1, n)
            return heap

        def heapDelete(heap, n, i):
            swap(heap, i, n-1)
            parent = (i-1)//2
            if heap[i] < heap[parent] or parent <= 0:
                heapify(heap, n-1, i)
            else:
                upMove(heap, n-1, i)
            heap.pop()
 
        window = nums[:k]
        heap = buildHeap(window, k)
        result = [heap[0]]
        
        i = 0
        while i < len(nums) - k:
            target = heap.index(nums[i])
            heapDelete(heap, k, target)
            heapInsert(heap, k-1, nums[i+k])
            result.append(heap[0])
            i += 1
        return result
            
```
