```python
class Solution(object):
    def isMatch(self, s, p):
        """
        :type s: str
        :type p: str
        :rtype: bool
        """
        
        def match(s, p):
            if s == p: return True
            elif p == "*": return True
            elif p == "?":
                return True if len(s) == 1  else False
            else:
                return True if s == p else False
                
        def recur(s, p):
            if (s, p) in mem: return mem[(s, p)]
            if p == "": return True if s == "" else False
            if s == "": return True if p == "" or set(p) == set(["*"]) else False
            if match(s, p):
                mem[(s, p)] = True
                return True
            
            res = set([False])
            if p[0] == "*":
                res = res | set([recur(s, p[1:]), recur(s[1:], p[1:]), recur(s[1:], p)])
            else:
                if match(s[0], p[0]):
                    res.add(recur(s[1:], p[1:]))
                else:
                    mem[(s, p)] = False
                    return False
            
            if True in res:
                mem[(s, p)] = True
            else:
                mem[(s, p)] = False
            return mem[(s, p)]
            
        mem = {}
        return recur(s, p)
    
```

```python
// golang
func match(s string, p string) bool {
    if s == p {
        return true
    } else if p == "*" {
        return true
    } else if p == "?" {
        if len(s) == 1 {
            return true
        } else {
            return false
        }
    } else {
        if s == p {
            return true
        } else {
            return false
        }
    }
    
}

func equalToStar(p string) bool {
    for _,e := range(p) {
        if string(e) != "*" {
            return false
        }
    }
    return true
}

func recur(s string, p string, mem map[[2]string] bool) bool {
    if val, ok := mem[[2]string{s, p}]; ok {
        return val
    }
    if p == "" {
        if s == "" {
            return true
        } else {
            return false
        }
    }
    if s == "" {
        if p == "" || equalToStar(p) {
            return true
        } else {
            return false
        }
    }
    if match(s, p) {
        mem[[2]string{s, p}] = true
        return true
    }
    
    res := map[bool] int { false:0 }
    if string(p[0]) == "*" {
        res[recur(s[1:], p[1:], mem)] = 0
        res[recur(s, p[1:], mem)] = 0
        res[recur(s[1:], p, mem)] = 0
    } else {
        if match(string(s[0]), string(p[0])) {
            res[recur(s[1:], p[1:], mem)] = 0
        } else {
            mem[[2]string{s, p}] = false
            return false
        }
    }
    if _, ok := res[true]; ok {
        mem[[2]string{s, p}] = true
    } else {
        mem[[2]string{s, p}] = false
    }
    return mem[[2]string{s, p}]
}

func isMatch(s string, p string) bool {
    mem := map[[2]string] bool {}
    return recur(s, p, mem)
}
```
