```python
// go
func reverse(nums []int) []int {
    res := []int{}
    for j := len(nums)-1; j >= 0; j-- {
        res = append(res, nums[j])
    }
    return res
}

func recur(n int, mem map[int]int, sqs []int) int {
    if val, ok := mem[n]; ok {
        return val
    }
    if n < 0 {
        return -1
    }
    if n == 0 {
        return 0
    }
    maxInt := 9999999
    
    minLen := maxInt
    for _,i := range(sqs) {
        subLen := recur(n-i, mem, sqs)
        if subLen != -1 && subLen < minLen {
            minLen = subLen
        }
    }
    
    if minLen == maxInt {
        mem[n] = -1
        return -1
    } else {
        mem[n] = minLen + 1
        return minLen + 1
    }
}

func numSquares(n int) int {
    sqs := []int{}
    for i := 1; i < n+1; i++ {
        if i * i <= n {
            sqs = append(sqs, i * i)
        } else {
            break
        }
    }
    
    sqs = reverse(sqs)
    
    mem := map[int] int {}
    return recur(n, mem, sqs)
    
}
```
