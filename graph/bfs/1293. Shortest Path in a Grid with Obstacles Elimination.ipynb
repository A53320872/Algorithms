```python
class Solution(object):
    def shortestPath(self, grid, k):
        """
        :type grid: List[List[int]]
        :type k: int
        :rtype: int
        """
        if len(grid) == 0: return -1
        row, col = len(grid), len(grid[0])
        
        start = (0,0,k)
        level = {start:0}
        queue = [start]
        visited = set([start])
        obs = {(0,0):k} # mark minimun obstacles to reduce total running time
        l = 1
        while queue:
            next_queue = []
            for node in queue:
                i,j,o = node[0], node[1], node[2]
                for nextNode in [(i+1, j, o), (i, j+1, o), (i-1, j, o), (i, j-1, o)]:
                    next_i, next_j = nextNode[0], nextNode[1]
                    if next_i >= 0 and next_j >=0 and next_i < row and next_j < col and nextNode not in visited:
                        if grid[next_i][next_j] == 0:
                            visited.add(nextNode)
                            next_queue.append(nextNode)
                            level[nextNode] = l
                        else:
                            if o <= 0 or ((next_i, next_j) in obs and obs[(next_i, next_j)] >= o):
                                continue
                            visited.add((next_i, next_j, o-1))
                            next_queue.append((next_i, next_j, o-1))
                            obs[(next_i, next_j)] = o-1
                            level[nextNode] = l
                            
            l += 1
            queue = next_queue
        # print(level)
        path = []
        for e in level:
            if e[0] == row-1 and e[1] == col - 1:
                path.append(level[e])
        return min(path) if path != [] else -1                 
```

```python
// golang
func min(nums []int) int {
    res := nums[0]
    for _,e := range(nums) {
        if e < res { res = e }
    }
    return res
}

func shortestPath(grid [][]int, k int) int {
    if len(grid) == 0 { return -1 }
    row, col := len(grid), len(grid[0])
    
    start := [3]int{0,0,k}
    level := map[[3]int]int{start:0}
    queue := [][3]int{start}
    visited := map[[3]int]bool{}
    obs := map[[2]int]int{[2]int{0,0}: k}
    l := 1
    for len(queue) > 0 {
        next_queue := [][3]int{}
        for _,node := range(queue) {
            i,j,o := node[0], node[1], node[2]
            for _,nextNode := range([][3]int{[3]int{i+1,j,o}, [3]int{i,j+1,o}, [3]int{i-1,j,o}, [3]int{i,j-1,o}}) {
                next_i, next_j := nextNode[0], nextNode[1]
                _,ok1 := visited[nextNode]
                if next_i >= 0 && next_j >=0 && next_i < row && next_j < col && !ok1 {
                    if grid[next_i][next_j] == 0 {
                        visited[nextNode] = true
                        next_queue = append(next_queue, nextNode)
                        level[nextNode] = l
                    } else {
                        val2,ok2 := obs[[2]int{next_i, next_j}]
                        if o <= 0 || (ok2 && val2 >= o) { continue }
                        visited[[3]int{next_i, next_j, o-1}] = true
                        next_queue = append(next_queue, [3]int{next_i, next_j, o-1})
                        obs[[2]int{next_i, next_j}] = o-1
                        level[nextNode] = l
                    }
                }
            } 
        }
        l += 1
        queue = next_queue
    }
    path := []int{}
    for k,v := range(level) {
        if k[0] == row-1 && k[1] == col-1 {
            path = append(path, v)
        }
    }
    if len(path) == 0 {
        return -1
    } else { return min(path) }
}
```
