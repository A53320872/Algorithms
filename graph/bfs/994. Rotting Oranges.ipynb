```python
# Regular BFS solution
# Just be careful about the edge cases
class Solution(object):
    def orangesRotting(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        row, col = len(grid), len(grid[0])
        i = 0
        rots = []
        numOfGood = 0
        while i < row:
            j = 0
            while j < col:
                if grid[i][j] == 2:
                    rots.append((i,j))
                elif grid[i][j] == 1:
                    numOfGood += 1
                j += 1
            i += 1
        
        if rots == []: return 0 if numOfGood == 0 else -1
        
        queue = rots
        level = {}
        for e in queue:
            level[e] = 0
        layer = 1
        while queue:
            next = []
            for u in queue:
                i, j = u[0], u[1]
                if i+1 < row:
                    if grid[i+1][j] == 1 and grid[i+1][j] not in level:
                        grid[i+1][j] = 2
                        level[(i+1, j)] = layer
                        next.append((i+1, j))
                if j+1 < col:
                    if grid[i][j+1] == 1 and grid[i][j+1] not in level:
                        grid[i][j+1] = 2
                        level[(i, j+1)] = layer
                        next.append((i, j+1))
                if i-1 >= 0:
                    if grid[i-1][j] == 1 and grid[i-1][j] not in level:
                        grid[i-1][j] = 2
                        level[(i-1, j)] = layer
                        next.append((i-1, j))
                if j-1 >= 0:
                    if grid[i][j-1] == 1 and grid[i][j-1] not in level:
                        grid[i][j-1] = 2
                        level[(i, j-1)] = layer
                        next.append((i, j-1))
            queue = next
            layer += 1
        i, j = 0, 0
        while i < row:
            j = 0
            while j < col:
                if grid[i][j] == 1:
                    return -1
                j += 1
            i += 1
        return max(level.values())
        
            
                    
                    
```
