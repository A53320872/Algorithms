```python
class Solution(object):
    def largestIsland(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        if len(grid) == 0: return 0
        row, col = len(grid), len(grid[0])
        
        def dfs(i,j, visited, c, mem):
            visited.add((i,j))
            mem[(i,j)] = c
            for next_node in [(i+1, j), (i, j+1), (i-1, j), (i, j-1)]:
                next_i, next_j = next_node
                if next_i < 0 or next_i >= row or next_j < 0 or next_j >= col \
                    or grid[next_i][next_j] == 0 or (next_i, next_j) in visited:
                    continue
                dfs(next_i, next_j, visited, c, mem)
                
            
        mem = {}
        comp = 0
        compLen = {}
        visited = set([])
        for i in range(row):
            for j in range(col):
                if grid[i][j] == 1 and (i,j) not in visited:
                    dfs(i,j, visited, comp, mem)
                    comp += 1
        # print('Components: ', mem)
        for k in mem:
            if mem[k] not in compLen:
                compLen[mem[k]] = 1
            else:
                compLen[mem[k]] += 1
        # print('CompLen: ', compLen)
        maxNum = 0
        for i in range(row):
            for j in range(col):
                if grid[i][j] == 0:
                    totalNum = 0
                    totalComp = set([])
                    for next_node in [(i+1, j), (i, j+1), (i-1, j), (i, j-1)]:
                        next_i, next_j = next_node
                        if next_i < 0 or next_i >= row or next_j < 0 or next_j >= col \
                           or grid[next_i][next_j] == 0:
                            continue
                        totalComp.add(mem[next_node])
                    for c in totalComp:
                        totalNum += compLen[c]
                    maxNum = max(maxNum, totalNum+1)
                    
        return maxNum if len(visited) != row * col else row * col # edge case
```

```python
// golang
func dfs(i int, j int, visited map[[2]int]bool, c int, mem map[[2]int]int, row int, col int, grid [][]int) {
    p := [2]int{i,j}
    visited[p] = true
    mem[p] = c
    
    for _,next_node := range([4][2]int{[2]int{i+1, j}, [2]int{i, j+1}, [2]int{i-1, j}, [2]int{i, j-1}}) {
        next_i, next_j := next_node[0], next_node[1]
        _,ok := visited[[2]int{next_i, next_j}]
        if next_i < 0 || next_i >= row || next_j < 0 || next_j >= col || grid[next_i][next_j] == 0 || ok { continue }
        dfs(next_i, next_j, visited, c, mem, row, col, grid)
    }
}

func largestIsland(grid [][]int) int {
    if len(grid) == 0 { return 0 }
    row, col := len(grid), len(grid[0])
    
    mem := map[[2]int]int{}
    comp := 0
    compLen := map[int]int{}
    visited := map[[2]int]bool{}
    
    for i := 0; i < row; i++ {
        for j:= 0; j < col; j ++ {
            p := [2]int{i,j}
            if grid[i][j] == 1 {
                if _,ok := visited[p]; !ok {
                    dfs(i, j, visited, comp, mem, row, col, grid)
                    comp += 1
                }
            }
        }
    }
    // fmt.Println("Components: ", mem)
    for k,_ := range(mem) {
        if _,ok := mem[k]; !ok {
            compLen[mem[k]] = 1
        } else {
            compLen[mem[k]] += 1
        }
    }
    // fmt.Println("CmpLen: ", compLen)
    maxNum := 0
    for i := 0; i < row; i++ {
        for j := 0; j < col; j ++ {
            if grid[i][j] == 0 {
                totalNum := 0
                totalComp := map[int]bool{}
                for _,next_node := range([4][2]int{[2]int{i+1, j}, [2]int{i, j+1}, [2]int{i-1, j}, [2]int{i, j-1}}) {
                    next_i, next_j := next_node[0], next_node[1]
                    if next_i < 0 || next_i >= row || next_j < 0 || next_j >= col || grid[next_i][next_j] == 0 { continue }
                    totalComp[mem[next_node]] = true
                }
                for c,_ := range(totalComp) {
                    totalNum += compLen[c]
                }
                if totalNum + 1 > maxNum {
                    maxNum = totalNum + 1
                } 
            }
        }
    }
    
    if len(visited) != row * col {
        return maxNum
    } else { return row * col }
    
}
```
