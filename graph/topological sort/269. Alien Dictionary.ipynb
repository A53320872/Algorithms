拓扑排序思想

1.建图（V， E， adjList， indegreeMap， outdegreeMap）
2.拓扑排序（kahn）算法
3.输出结果

```python
class Solution(object):
    def alienOrder(self, words):
        """
        :type words: List[str]
        :rtype: str
        """
        # Build graph
        def buildGraph(words):
            V = getVerticesSet(words)
            E = getEdgesSet(words)
            [indeg, outdeg] = getDegreeMap(V, E)
            
            adjList = getAdjMap(V, E)
            return [V, E, adjList, indeg, outdeg]
        
        # Vertices set
        def getVerticesSet(words):
            V = set()
            for word in words:
                for e in word:
                    if e not in V:
                        V.add(e)
            return V
        
        # Edges set
        def getEdgesSet(words):
        
            E = set()

            # considering adjacent words
            i = 1
            while i < len(words):
                
                cur = words[i-1]
                nxt = words[i]
                
                i += 1
            
                index = getFirstDiffPos(cur, nxt)
                if index == -1 and len(cur) <= len(nxt): 
                    continue
                
                # special invalid case e.g.('ab', 'a')
                elif index == -1 and len(cur) > len(nxt):
                    return False
                
                currword = cur[index]
                nextword = nxt[index]

                edge = (currword, nextword)
                if edge not in E:
                    E.add(edge)
                
            return E
        
        # Get first different elemnt's position
        def getFirstDiffPos(cur, nxt):
            mini = len(cur) if len(cur) < len(nxt) else len(nxt)

            i = 0
            while i < mini:
                if cur[i] != nxt[i]:
                    return i

                i += 1
            return -1
        
        def getDegreeMap(V, E):
            indegMap = {}
            outdegMap = {}
            for v in V:
                indegMap[v] = 0
                outdegMap[v] = 0
            
            # left -> right
            for edge in E:
                left = edge[0]
                right = edge[1]
                
                if left not in outdegMap:
                    outdegMap[left] = 1
                else:
                    outdegMap[left] += 1
                
                if right not in indegMap:
                    indegMap[right] = 1
                else:
                    indegMap[right] += 1
                    
            return [indegMap, outdegMap]
        
        def getTopologicalSort(V, E, indegMap, outdegMap, adjMap):
            result = []
            zeroIndegSet = getZeroIndegSet(indegMap)
            print(zeroIndegSet)
            if zeroIndegSet == set():
                return False
            
            while zeroIndegSet:
                print(zeroIndegSet)
                node = getOneNode(zeroIndegSet)
                zeroIndegSet = removeNode(zeroIndegSet, node)
                result.append(node)
                for e in adjMap[node]:
                    edge = (node, e)
                    
                    E.remove(edge)
                    
                    indegMap[e] -= 1
                    if indegMap[e] == 0: 
                        zeroIndegSet.add(e)
            return result
                    
                    
                    
        def removeNode(zeroIndegSet, node):
            zeroIndegSet.remove(node)
            return zeroIndegSet if zeroIndegSet != None else set()
                    
                    
                
        def getAdjMap(V, E):
            adjMap = {}
            for v in V:
                adjMap[v] = set()
            
            for e in E:
                left = e[0]
                right = e[1]
                
                adjMap[left].add(right)
            
            return adjMap
        
        def getOneNode(zeroIndegSet):
            for e in zeroIndegSet:
                return e
            return False
            
        def getZeroIndegSet(indegMap):
            zeroIndegSet = set()
            for e in indegMap:
                if indegMap[e] == 0:
                    zeroIndegSet.add(e)
            return zeroIndegSet
        
        if getEdgesSet(words) == False: return ""
        [V, E, adjMap, indegMap, outdegMap] = buildGraph(words)
        toposort = getTopologicalSort(V, E, indegMap, outdegMap, adjMap)
        return "".join(toposort) if (toposort != False and len(toposort) == len(V)) else ""
        
        
        
        
                
```
