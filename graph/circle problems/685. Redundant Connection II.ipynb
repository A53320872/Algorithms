```python
class Solution(object):
    def findRedundantDirectedConnection(self, edges):
        """
        :type edges: List[List[int]]
        :rtype: List[int]
        """
        indg = {}
        adj = {}
        pa = {}
        for (u,v) in edges:
            adj[u] = set([])
            adj[v] = set([])
            indg[u] = 0
            indg[v] = 0
        for (u,v) in edges:
            pa[u] = u
            pa[v] = v
            adj[u].add(v)
            indg[v] += 1
        
        start = -1
        for node in indg:
            if indg[node] == 0:
                start = node
                break
        
            
        def union(u, v):
            pa[u] = v
            
        def find(u):
            while pa[u] != u:
                u = pa[u]
            return u
        
        res = [[]]
        
        def dfs(u, visited):
            visited.add(u)
            for v in adj[u]:
                if v in visited:
                    res[0] = [u, v]
                    continue
                dfs(v, visited)
        
        for u, v in edges:
            if pa[v] == v:
                pa[v] = u
            
            # multi parents
            else:
                e1 = [pa[v], v]
                e2 = [u, v]
                
                # print(e1, e2, v)
                dfs(v, set([]))
                
                # no cicle
                if res[0] == []:
                    for (x,y) in edges[::-1]:
                        if x == e1[0] and y == e2[0]: return e1
                        else: return e2
                # has cicle
                else:
                    return res[0]
        
        # no multi parent node, just find cicle (same as undirected graph)
        # reset parents map
        for (u,v) in edges:
            pa[u] = u
            pa[v] = v
        res = []
        for (u, v) in edges:
            p_u = find(u)
            p_v = find(v)
            if p_u != p_v:
                union(p_u, p_v)
            else:
                return [u, v]
        
        
        
        
                
                
                
            
                
        
```
