```python
"""
# Definition for an Interval.
class Interval(object):
    def __init__(self, start=None, end=None):
        self.start = start
        self.end = end
"""
## 1st method
## compute free intervals, then intersect them
class Solution(object):
    def employeeFreeTime(self, schedule):
        """
        :type schedule: [[Interval]]
        :rtype: [Interval]
        """
        def computeFree(A):
            freeArray = []
            lastEnd = float("-inf")
            for i, e in enumerate(A):
                if i == 0:
                    freeArray.append([float("-inf"), e.start])
                    lastEnd = e.end
                if i != 0:
                    freeArray.append([lastEnd, e.start])
                    lastEnd = e.end
                if i == len(A) - 1:
                    freeArray.append([e.end, float("inf")])
                    break
                    
            return freeArray
        
        def interSect(A, B):
            res = []
            i, j = 0, 0
            while i < len(A) and j < len(B):
                lo = max(A[i][0], B[j][0])
                hi = min(A[i][1], B[j][1])

                if lo < hi:
                    res.append([lo, hi])

                if A[i][1] < B[j][1]:
                    i += 1
                else:
                    j += 1
            return res
                
        target = [[float("-inf"), float("inf")]]
        for e in schedule:
            target = interSect(target, computeFree(e))
        
        res = []
        for e in target:
            if e[0] in [float("-inf"), float("inf")] or e[1] in [float("-inf"), float("inf")]:
                continue
            else:
                a = Interval(None)
                a.start, a.end = e[0], e[1]
                res.append(a)
        
        return res
            
```

```python
"""
# Definition for an Interval.
class Interval(object):
    def __init__(self, start=None, end=None):
        self.start = start
        self.end = end
"""
## 2nd method (nlogn)
## merge intervals, then compute free intervals
class Solution(object):
    def employeeFreeTime(self, schedule):
        """
        :type schedule: [[Interval]]
        :rtype: [Interval]
        """
        def createInterval(interval):
            new = Interval(None)
            new.start, new.end = interval[0], interval[1]
            return new
        
        def computeFree(A):
            freeArray = []
            lastEnd = float("-inf")
            for i, e in enumerate(A):
                if i == 0:
                    freeArray.append(createInterval([float("-inf"), e.start]))
                    lastEnd = e.end
                if i != 0:
                    freeArray.append(createInterval([lastEnd, e.start]))
                    lastEnd = e.end
                if i == len(A) - 1:
                    freeArray.append(createInterval([e.end, float("inf")]))
                    break
                    
            return freeArray
        
        def merge(sortedArrays):
            stack = [sortedArrays[0]]
            for i in range(1, len(sortedArrays)):
                
                cur = sortedArrays[i]
                top = stack[len(stack) - 1]
                
                if cur.start > top.end:
                    stack.append(cur)
                
                elif cur.end < top.end:
                    continue
                
                else:
                    stack.pop()
                    new = Interval(None)
                    new.start, new.end = top.start, cur.end
                    stack.append(new)
                    
            return stack
        
        combinedArray = []
        for e in schedule:
            combinedArray += e
        combinedArray.sort(key=lambda x: x.start)
        
        res = []
        for e in computeFree(merge(combinedArray)):
            if e.start not in [float("-inf"), float("inf")] and e.end not in [float("-inf"), float("inf")]:
                res.append(e)
    
        return res
            
```
